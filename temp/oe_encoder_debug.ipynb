```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import pickle
from sklearn.model_selection import train_test_split
```

```python
encoder_oe = pickle.load(open("./artifacts/encoder_oe_.pkl", "rb"))
```

```python
CAT_COLS_OE = [
    "OnlineSecurity",
    "MultipleLines",
    "Dependents",
    "DeviceProtection",
    "StreamingTV",
    "OnlineBackup",
    "gender",
    "SeniorCitizen",
    "PhoneService",
    "Partner",
    "PaperlessBilling",
    "StreamingMovies",
    "TechSupport",
]
```

```python
data = pd.read_csv("./../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
data
```

```python
TARGET = ["Churn"]
X, y = data.drop(columns=TARGET), data[TARGET]
X_train, X_test, y_train, y_test = train_test_split(X, y)
type(y_train)
```

```python
encoder_oe_new = OrdinalEncoder()
```

```python
X_train.shape
```

```python
X_oe__train = X_train[CAT_COLS_OE]
print(X_oe__train.shape)
encoder_oe_new.fit(X_oe__train)
X_oe_trans__train = encoder_oe_new.transform(X_oe__train)
print(X_oe_trans__train.shape)
```

```python
X_oe_trans__train_df = pd.DataFrame(X_oe_trans__train, columns=CAT_COLS_OE)
print(X_oe_trans__train_df.shape)
print(type(X_oe_trans__train_df))
```
